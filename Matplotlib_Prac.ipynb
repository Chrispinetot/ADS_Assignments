# Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

** Import matplotlib.pyplot as plt and set %matplotlib inline if you are using the jupyter notebook. What command do you use if you aren't using the jupyter notebook?**

```python
#If not in jupyter notebook you use plt.show()
%matplotlib inline
from matplotlib import pyplot as plt
```

## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
#Creating a figure object
figure = plt.figure()

#Creating axis
ax = figure.add_axes([0,0,1,1])

#xy plot
ax.plot(x,y, "b")
ax.set_title("title")
ax.set_ylabel("y")
ax.set_xlabel("x")
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
#creating figure object
figure_obj = plt.figure()

#adding axis 1 and 2 to figure object
axis_1 = figure_obj.add_axes([0,0,1,1])
axis_2 = figure_obj.add_axes([.2,.5,.4,.4])

#Plots
axis_1.plot(x,z)
axis_1.set_xlabel("X")
axis_1.set_ylabel("Z")
axis_2.plot(x,y)
axis_2.set_xlabel("X")
axis_2.set_ylabel("Y")
plt.xlim(20,22)
plt.ylim(30,50)
axis_2.set_title("zoom")
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
#Method 1
fig, (ax_1,ax_2) = plt.subplots(1,2)
ax_1.plot(x,y, "b", linestyle="dashed", linewidth=4,)
ax_2.plot(x,z, 'r', linewidth=4)
```

```python
#Method 2
fig = plt.figure()
ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(x,y, "b", linestyle="dashed", linewidth=4,)
ax_2.plot(x,z, 'r', linewidth=4)
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
#Method 1
fig, (ax_1,ax_2) = plt.subplots(1,2, figsize = (20,8))

ax_1.plot(x,y, "b", linestyle="dashed", linewidth=4,)
ax_2.plot(x,z, 'r', linestyle="dotted", linewidth=4)
```

```python
fig = plt.figure(figsize=(20,8))
ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(x,y, "b", linestyle="dashed", linewidth=4,)
ax_2.plot(x,z, 'r', linestyle="dotted", linewidth=4)
```
